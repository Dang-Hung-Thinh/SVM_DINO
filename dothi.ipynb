```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1 = "mpu6050_data.csv"
file2 = "mpu6050_data_jmp.csv"
```

```python
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Vẽ dữ liệu từ file 1
axes[0].plot(df1["Time(ms)"], df1["AccX"], label="AccX", color="red")
axes[0].plot(df1["Time(ms)"], df1["AccY"], label="AccY", color="green")
axes[0].plot(df1["Time(ms)"], df1["AccZ"], label="AccZ", color="blue")
axes[0].set_title("Dữ liệu từ file 1")
axes[0].set_xlabel("Time (ms)")
axes[0].set_ylabel("Acceleration")
axes[0].legend()
axes[0].grid()

# Vẽ dữ liệu từ file 2
axes[1].plot(df2["Time(ms)"], df2["AccX"], label="AccX", color="red")
axes[1].plot(df2["Time(ms)"], df2["AccY"], label="AccY", color="green")
axes[1].plot(df2["Time(ms)"], df2["AccZ"], label="AccZ", color="blue")
axes[1].set_title("Dữ liệu từ file 2")
axes[1].set_xlabel("Time (ms)")
axes[1].set_ylabel("Acceleration")
axes[1].legend()
axes[1].grid()

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
```
